```python
X=df.drop('R',axis=1)
y=df['R']
```

```python
y_scaled=df['R'].replace({'R':0,'M':1})
```

```python
print(y_scaled.unique())
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y_scaled,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')

])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp=model.predict(X_test)
yp_binary=(yp>0.5).astype(int)
print(yp_binary[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,yp_binary))
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')

])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_train,y_train)
```

****

```python
y_prediction=model.predict(X_test)
y_prediction_b=(y_prediction>0.5).astype(int)
y_prediction_b[:10]
```

```python
y_test[:10]
```

```python
print(classification_report(y_test,y_prediction_b))
```
